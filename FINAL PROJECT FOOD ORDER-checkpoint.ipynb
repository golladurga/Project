```python
class Admin:
    def __init__(self):
        self.food_items = []
        self.food_id_counter = 1

    def add_food_item(self, name, quantity, price, discount, stock):
        food_id = self.food_id_counter
        self.food_id_counter += 1
        food_item = {
            'FoodID': food_id,
            'Name': name,
            'Quantity': quantity,
            'Price': price,
            'Discount': discount,
            'Stock': stock
        }
        self.food_items.append(food_item)
        print(f"Food item with FoodID {food_id} added successfully.")

    def edit_food_item(self, food_id, new_data):
        for food_item in self.food_items:
            if food_item['FoodID'] == food_id:
                food_item.update(new_data)
                print(f"Food item with FoodID {food_id} edited successfully.")
                return
        print(f"Food item with FoodID {food_id} not found.")

    def view_food_items(self):
        for food_item in self.food_items:
            print(food_item)

    def remove_food_item(self, food_id):
        for food_item in self.food_items:
            if food_item['FoodID'] == food_id:
                self.food_items.remove(food_item)
                print(f"Food item with FoodID {food_id} removed successfully.")
                return
        print(f"Food item with FoodID {food_id} not found.")


# Example usage:
admin = Admin()
admin.add_food_item("Tandoori Chicken", "4 pieces", 240, 0, 10)
admin.add_food_item("Vegan Burger", "1 piece", 320, 0, 5)
admin.view_food_items()
admin.edit_food_item(1, {'Name': 'Tandoori Chicken (Full)', 'Price': 400})
admin.remove_food_item(2)

class User:
    def __init__(self):
        self.orders = []
        self.order_id_counter = 1

    def register(self, full_name, phone_number, email, address, password):
        # Implement registration logic
        print("User registered successfully.")

    def login(self, email, password):
        # Implement login logic
        print("User logged in successfully.")

    def place_new_order(self, food_items):
        # Implement order placement logic
        order_id = self.order_id_counter
        self.order_id_counter += 1
        order = {
            'OrderID': order_id,
            'FoodItems': food_items
        }
        self.orders.append(order)
        print(f"Order with OrderID {order_id} placed successfully.")

    def order_history(self):
        for order in self.orders:
            print(order)

    def update_profile(self, new_data):
        # Implement profile update logic
        print("Profile updated successfully.")


# Example usage:
user = User()
user.register("John Doe", "1234567890", "john.doe@example.com", "123 Main St", "password")
user.login("john.doe@example.com", "password")
user.place_new_order([2, 3])
user.order_history()
user.update_profile({'Phone Number': '9876543210'})
```
